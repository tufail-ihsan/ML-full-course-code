**14. Convertors**

```python
# when it want to change some values name with other
# here i want to change Royal Challengers Bangalore into RCB
def rename(name):
    if name=="Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('matches - matches.csv',converters={'team1':rename,'team2':rename}).head()
```
