**Feature scaling**

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Social_Network_Ads.csv',usecols=['Age','EstimatedSalary','Purchased'])
```

```python
df.sample(3)
```

Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df.drop('Purchased',axis=1),
                                                df['Purchased'],
                                                test_size=0.3,
                                                random_state=0)
x_train.shape, x_test.shape
```

standardscaler

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
# fit the scaler to the train set, it will learn the parameters (imp point)
scaler.fit(x_train)

# transform train and test sets
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
scaler.mean_
```

before scaling

x_train # in dataframe form

after scaling

```python
x_train_scaled # both are in different form one another, onr in data frame and another in array form
# only between -1 to 1
```

```python
# covert it into dataframe

x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns) 
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_train.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")

ax2.scatter(x_train_scaled['Age'],
            x_train_scaled['EstimatedSalary'],
            color='red')
ax2.set_title("After Scaling")

plt.show()
```

```python
fig, ax2 = plt.subplots(figsize=(6, 4))

ax2.set_title('After Standard Scaling')

sns.kdeplot(data=x_train_scaled, x='Age', ax=ax2, label='Age')
sns.kdeplot(data=x_train_scaled, x='EstimatedSalary', ax=ax2, label='EstimatedSalary')

ax2.legend()

plt.show()
print(x_train_scaled['Age'].nunique())
print(x_train_scaled['EstimatedSalary'].nunique())
```

```python
fig, (ax1, ax2)=plt.subplots(ncols=2, figsize=(12,5))

#before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(x_train['Age'],ax=ax1)

#After scaling
ax2.set_title('Age Distribution After Scaling')
sns.kdeplot(x_train_scaled['Age'],ax=ax2)
plt.show()
```
