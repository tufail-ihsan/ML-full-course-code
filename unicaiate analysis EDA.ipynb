```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head(3)
```

```python
df.info()
```

**1. Categorical Data**

a.countplot

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
sns.countplot(df['Embarked'])
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

**b. PieChart**

```python
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```

**2. Numerical Data**

A. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=10) # bins can be change
```

b. Displot

```python
#improvement of histogram
# tell us about schevness of data
sns.distplot(df['Age'])
# KDE: kernal dinsity estimation( line on graph above on histogram)
# PDF: probability density function (show distribution of data)
# this show the probability of that number.
```

c. Boxplot

```python
# give five number summery
# min= (Q1-1.5*IQR)  IQR= Q3-Q1
# max= (Q3+1.5*IQR)
# Q1--> 25th percentile (percentile mean: 25% value is less than Q1 value)
# Q2--> 75th percentile(percentile mean: 75% value is less than Q1 value)

sns.boxplot(df['Age'])
```
