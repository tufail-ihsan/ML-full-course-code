```python
import requests
import pandas as pd

url = "https://movies-ratings2.p.rapidapi.com/ratings"

headers = {
    "x-rapidapi-key": "b35ebbf597mshcd8fed239e5035ap1016f7jsnbc5d586aa885",
    "x-rapidapi-host": "movies-ratings2.p.rapidapi.com",
    "Content-Type": "application/json"
}

movie_ids = [
    "tt0111161",  # Shawshank Redemption
    "tt0068646",  # The Godfather
    "tt0468569",  # The Dark Knight
    "tt0109830",  # Forrest Gump
    "tt0133093"   # The Matrix
]

movies = []

for movie_id in movie_ids:
    response = requests.get(
        url,
        headers=headers,
        params={"id": movie_id}
    )

    data = response.json()

    movies.append({
        "IMDb_ID": data.get("imdbId"),
        "TMDB_ID": data.get("tmdbId"),
        "IMDb_Score": data.get("ratings", {})
                           .get("imdb", {})
                           .get("score"),
        "Reviews_Count": data.get("ratings", {})
                              .get("imdb", {})
                              .get("reviewsCount")
    })

df = pd.DataFrame(movies)

df
```

```python
df.to_csv("movies_dataset.csv", index=False)

print("Dataset saved successfully!")
```
