**1. One Hot Encoding using pandas**

```python
pd.get_dummies(df,columns=['Fuel Types'])
#the fuel Types column is converted into 4 columns with 0 and 1 values. 
# This is called one hot encoding.
#each column represents a category and the value is 1 if the row belongs to that category and 0 otherwise.
```

**now**, here we will solve multicolinerity, to remove one column from fule tye colum

```python
pd.get_dummies(df, columns=['Fuel Types'], drop_first=True)
# drop_first=True drops the first column to avoid multicollinearity.
```

**2. OneHotEncoding using sklearn**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['Company Names','Fuel Types']],df['Cars Prices'],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first', sparse_output=False,dtype=np.int32)
#drop='first' drops the first column to avoid multicollinearity.
#sparse=False returns a dense array instead of a sparse matrix.
#dtype=np.int32 returns the array as int32 instead of float64.
```

on which colum we do encoding we will separate these colum and perform encoding after encoding we will again add with another columns, but htis is hard method. 

we will do this with **column transformer**, this is easy method.

```python
x_train_new=ohe.fit_transform(x_train[['Fuel Types']])
```

```python
x_test_new=ohe.transform(x_test[['Fuel Types']])
```

```python
x_train_new
```

now we cut the all the colums from datset except 'Fule Types' and append with 'x_train_new' and 'x_test_new'

```python
np.hstack((x_train[['Company Names']].values,x_train_new))
```

```python
np.hstack((x_test[['Company Names']].values,x_test_new))
```
