**Encoding Categorical Data**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('customer.csv',usecols=['review','education','purchased'])
```

```python
#convert colum into lower case
df['review']=df['review'].str.lower()
df['education']=df['education'].str.lower()
df['purchased']=df['purchased'].str.lower()
```

only input and output columns of data

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['review','education']],df['purchased'],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

**ordinal encoding of input columns**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe= OrdinalEncoder(categories=[['poor','average','good','excellent'],['school','ug','pg']])
```

**fit on only train data but transform on both train and test data**

```python
oe.fit(x_train)
```

```python
x_train=oe.transform(x_train)
x_test=oe.transform(x_test)
```

x_train after encoding

```python
x_train
```

**Label Encoding for output colum**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
le.fit(y_train)
```

```python
y_train=le.transform(y_train)
y_test=le.transform(y_test)
```
